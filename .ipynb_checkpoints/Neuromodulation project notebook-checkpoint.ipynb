# First attempt at connecting V1 cells
4 neuron types (SST, PV, VIP, pyramidal) are modelled as leaky integrate-and-fire (LIF) neurons and connected as in Pakan et. al (2016) using Brian2.

```python
from brian2 import *
import numpy
prefs.codegen.target="numpy"

eqs = '''
dy/dt = -y/tau_syn : 1
dx/dt = (y - x)/tau_syn: 1
dv/dt = (v0-v)/(R*C) + I/C - x*(v-v_syn)/(R*C): volt (unless refractory)
v0 : volt
I : amp
refperiod : second
C : farad
R : ohm
vt : volt
v_syn : volt
tau_syn : second
'''
# Neurons in order: VIP, SST, PV, pyramidal
V1 = NeuronGroup(4, eqs, threshold='v>vt', reset='v = 0*mV', refractory='refperiod', method='euler')
V1.I = [70, 50, 70, 150]*pA # Point of mild activity
V1.refperiod = [2.4,2.4,2.4,4.65]*ms
V1.C = [40.31,40.31,40.31,75.492]*pfarad
V1.R = [525,525,525,281]*Mohm
V1.vt = [22.92,22.92,22.92,26.26]*mV
V1.tau_syn = [10,10,10,10]*ms
```

Two cells, a layer 4 pyramidal cell and an SST cell, were found modelled as a LIF neuron in the Allen Cell Types Database. The typical parameters were used to implement an LIF model in Brian2, using the SST cell parameters for all inhibitory cells and pyramidal cell parameters for all excitatory ones. Currents were chosen to make sure all cells were active, but barely. Equations for an alpha synapse were also included, and its time constant and reversal potential set to randomly chosen plausible values.

```python
S=Synapses(V1,V1,'w : 1', on_pre='y += w')
S.connect(i=0, j=2)
S.connect(i=1, j=[0,2,3])
S.connect(i=2, j=[2,3])
S.w = 1

M = StateMonitor(V1, 'v', record=True)

run(200*ms)

plot(M.t/ms, M.v[0], '-b', label='VIP')
plot(M.t/ms, M.v[1], '-g', label='SST')
plot(M.t/ms, M.v[2], '-c', label='PV')
plot(M.t/ms, M.v[3], '-y', lw=3, label='Pyramidal')
xlabel('Time (ms)')
ylabel('v')
legend(loc='best')
show()
```

Cells are connected as in connectivity diagram. Synaptic weights are all set to 1 (arbitrarily).
